# Multiple Linear Regression for Robot Calibration

In this lab, we will illustrate the use of multiple linear regression for calibrating robot control.  In addition to reviewing the concepts in the [multiple linear regression demo](./glucose.ipynb), you will see how to use multiple linear regression for time series data -- an important concept in dynamical systems such as robotics.

The robot data for the lab is taken generously from the TU Dortmund's [Multiple Link Robot Arms Project](http://www.rst.e-technik.tu-dortmund.de/cms/en/research/robotics/TUDOR_engl/index.html). We will focus on predicting the current draw into one of the joints as a function of the robot motion.  Such models are essential in predicting the overall robot power consumption.  Several other models could also be used.

Below, I have supplied the column headers in the `names` array.  Use the `pd.read_csv` command to load the training data in `exp1.csv`.  Use the `index_col` option to specify that column 0 (the one with time) is the *index* column.  You can review [simple linear regression demo](..\simp_lin_reg\auto_mpg.ipynb) for examples of using the `pd.read_csv` command.

```python
names =[
    't',                                  # Time (secs)
    'q1', 'q2', 'q3',                     # Joint angle   (rads)
    'dq1', 'dq2', 'dq3',                  # Joint velocity (rads/sec)
    'I1', 'I2', 'I3',                     # Motor current (A)
    'eps21', 'eps22', 'eps31', 'eps32',   # Strain gauge measurements ($\mu$m /m )
    'ddq1', 'ddq2', 'ddq3'                # Joint accelerations (rad/sec^2)
]
# TODO 1
# df = pd.read_csv(...)

df = pd.read_csv('exp1.csv', header=None,names=names,index_col=0)
 
```

Print the first six lines of the pandas dataframe and manually check that they match the first rows of the csv file.

```python
# TODO 2
df.head(6)
```

From the dataframe `df`, extract the time indices into a vector `t` and extract `I2`, the current into the second joint.  Place the current in a vector `y` and plot `y` vs. `t`.   Label the axes with the units.

```python
# TODO 3
# y = ...
# t = ...
# plt.plot(...)


t = df.index
y = np.array(df['I2'])

fig=plt.figure(figsize=(20, 15))
plt.plot(t,y,'o')
plt.xlabel('Time')
plt.ylabel('Current into the second joint')
plt.grid(True)
```

Use all the samples from the experiment 1 dataset to create the training data:
* `ytrain`:  A vector of all the samples from the `I2` column
* `Xtrain`:  A matrix of the data with the columns:  `['q2','dq2','eps21', 'eps22', 'eps31', 'eps32','ddq2']`

```python
# TODO 4
# ytrain = ...
# Xtrain = ...

ytrain = np.array(df['I2'])

Xtrain = np.array(df[['q2','dq2','eps21', 'eps22', 'eps31', 'eps32','ddq2']])



```

## Fit a Linear Model
Use the `sklearn.linear_model` module to create a `LinearRegression` class `regr`.

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression

# Create linear regression object
# TODO 5
# regr = ...

regr = linear_model.LinearRegression()
```

Train the model on the training data.

```python
# TODO 6
regr.fit(Xtrain,ytrain)
```

Using the trained model, compute, `ytrain_pred`, the predicted current.  Plot `ytrain_pred` vs. time `t`.  On the same plot, plot the actual current `ytrain` vs. time `t`.  Create a legend for the plot.

```python
# TODO 7
# ytrain_pred = ...
# plt.plot(...)
ytrain_pred =regr.predict(Xtrain)

fig=plt.figure(figsize=(20, 15))

plt.plot(t,ytrain_pred,'o')
plt.plot(t,ytrain,'o','r')
plt.xlabel('Time')
plt.ylabel('Current into the second joint')
plt.grid(True)
plt.title('Plot of actual current, predicted current and time')
plt.legend(['actual', 'predicted'])
```

Measure the normalized RSS given by `RSS / (n s^2_y).

```python
# TODO 8
# RSS_train = ...

RSS_train = np.mean((ytrain_pred-ytrain)**2)/(np.std(ytrain)**2)
Rsq_train = 1-RSS_train
print ("Normalised value of traning data = %f." %RSS_train)
print ("R square training data data = %f." %Rsq_train)
```

## Measure the Fit on an Indepdent Dataset

Up to now, we have only tested the model on the same data on which it was trained.  In general, we need to test model on independent data not used in the training.  For this purpose, load the data in `exp2.csv`.  Compute the regression predicted values on this data and plot the predicted and actual values over time.

```python
# TODO 9


dft = pd.read_csv('exp2.csv', header=None,names=names,)
index_col = 0

t_test = np.array(dft['t'])
y_test = np.array(dft['I2'])

ytest = np.array(dft['I2'])

Xtest = np.array(dft[['q2','dq2','eps21', 'eps22', 'eps31', 'eps32','ddq2']])

y_test_pred = regr.predict(Xtest)

fig=plt.figure(figsize=(20, 15))

plt.plot(t,ytest,'o')
plt.plot(t,y_test_pred,'o','r')
plt.xlabel('Time')
plt.ylabel('Current into the second joint')
plt.grid(True)
plt.title('Plot of actual current, predicted current and time of TEST DATA')
plt.legend(['actual', 'predicted'])









```

Measure the normalized RSS on the test data.

```python
# TODO 10
# RSS_test = ...

RSS_test = np.mean((y_test_pred-ytest)**2)/(np.std(ytest)**2)
Rsq_test = 1-RSS_test
print ("Normalised value of test data = %f." %RSS_test)
print ("R square test data = %f." %Rsq_test)
```
